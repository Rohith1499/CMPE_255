[LP 2] Use encoders and convert the categorical variables to numerical values: Columns A1, A4, A5, A6, A7, A9, A10, A12, A13

[HP 1] Describe your choice of encoder for these columns and explain why they are appropriate for the column.


```python
## Your code for LP 2 goes here.
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
data["Gender"] = le.fit_transform(data["Gender"])
data["PriorDefault"] = le.fit_transform(data["PriorDefault"])
data["Employed"] = le.fit_transform(data["Employed"])
data["DriversLicense"] = le.fit_transform(data["DriversLicense"])
data["Approved"] = le.fit_transform(data["Approved"])

from sklearn.preprocessing import OrdinalEncoder
oe= OrdinalEncoder()
data['Married'] = oe.fit_transform(data['Married'].to_numpy().reshape(-1, 1))
data['BankCustomer'] = oe.fit_transform(data['BankCustomer'].to_numpy().reshape(-1, 1))
data['EducationLevel'] = oe.fit_transform(data['EducationLevel'].to_numpy().reshape(-1, 1))
data['Ethnicity'] = oe.fit_transform(data['Ethnicity'].to_numpy().reshape(-1, 1))
data['Citizen'] = oe.fit_transform(data['Citizen'].to_numpy().reshape(-1, 1))
```

For the columns: Gender, PriorDefault, Employed, DriversLicense, Approved we have only binary values like yes/no, male/female so I have used Label encoding to transform those categorical features to 1's and 0's



For the columns: Married, BankCustomer, EducationLevel, Ethnicity, Citizen. I have used ordinal encoder instead of Label encoder and One hot encoder. Reason for not using label encoding because it coverts into 1's and 0's but we have many values for each feature so other values get nullfied or overlooked in the process. Reason for not using One Hot encoding because as said before there are many values in each of the above feature due to each values gets its own feature due to which dimensions increases and sparsity between the points increases which result in curse of dimensionality
